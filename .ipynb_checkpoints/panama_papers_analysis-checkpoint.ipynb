# "Worldwide company" Panama Papers Analysis. Potential risk areas

```python
import pandas
import seaborn
import matplotlib.pyplot as plot
```

```python
baltics = ["Estonia", "Latvia", "Lithuania"]
CIS_countries = ["Armenia", "Azerbaijan", "Belarus", "Kazakhstan", "Kyrgyzstan", "Moldova", "Russia", "Tajikistan", "Uzbekistan", "Turkmenistan"]
```

## Officers table

There are a lot of names in Panama Papers that can be potentially connected to "Worldwide Company" as well. Full list of them from baltic and CIS countries can be found in files 'officers_valuable_data.csv' and 'cis_officers_valuable_data.csv' respectively.

```python
officers_dataframe = pandas.read_csv("/home/school/Desktop/jupyter_files/panama_papers_analysis/officers_valuable_data.csv")
value_counts = officers_dataframe["countries"].value_counts()
plot.figure(figsize=(8, 6))
plot.bar(value_counts.index, value_counts.values, color="orange")
plot.xlabel("Baltic country")
plot.ylabel("Count")
plot.title("Count of officers in each baltic country")
plot.show()
```

```python
officers_dataframe = pandas.read_csv("/home/school/Desktop/jupyter_files/panama_papers_analysis/cis_officers_valuable_data.csv")
value_counts = officers_dataframe["countries"].value_counts()
plot.figure(figsize=(12, 6))
plot.bar(value_counts.index, value_counts.values, color="orange")
plot.xlabel("CIS country")
plot.ylabel("Count")
plot.yscale("log")
plot.title("Count of officers in each CIS country")
plot.show()
```
